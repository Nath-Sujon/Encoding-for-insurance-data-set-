# Encoding Techniques for Insurance Dataset
 Here we have used four type of encoding techniques
 (1)Without encoding (2)Level encoding (3) One-hot encoding (4) Ordinal encoding

```python
import pandas as pd
df = pd.read_csv('Insurance.csv')
```

```python
df.info()
```

```python
df.isnull().sum()
```

# (1) Without Encoding Techniques

```python
df1.head()
```

```python
df1.sex.unique()
```

```python
df1.sex = df1.sex.replace(['female', 'male'],[1,2])
```

```python
df1.head()
```

```python
df1.smoker.unique()
```

```python
df1.smoker = df1.smoker.replace(['yes', 'no'],[1,2])
```

```python
df1.head()
```

```python
df1.region.unique()
```

```python
df1.region = df1.region.replace(['southwest', 'southeast', 'northwest', 'northeast'],[1,2,3,4])
```

```python
df1.head()
```

```python
# Label Encoder using for Loop


import numpy as np
import warnings as wr
wr.filterwarnings('ignore')

columns = ['sex','smoker','region']
for column in df3.columns:
    if df3[column].dtype == np.number:
        continue
    df3[column] = le.fit_transform(df3[column])
```

```python
# Another way to encode using for loop

from pandas.core.dtypes.common import is_numeric_dtype
```

```python
for column in df4.columns:
    if is_numeric_dtype(df4[column]):
        continue
    else:
        df4[column] = le.fit_transform(df4[column])
```

# (4) Ordinal Encoding Technique

```python
df8.head()
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
gender = df8.sex.unique()
```

```python
gender
```

```python
ordinal = OrdinalEncoder(categories=[gender])
```

```python
ordinal
```

```python
encoded = ordinal.fit_transform(df8[['sex']])
```

```python
encoded
```

```python
en = pd.DataFrame(encoded,columns=['encoded_sex'])
```

```python
en.head()
```

```python
df8=df8.drop('sex',axis=1)   # Drop sex column
df8=pd.concat([df8,en],axis=1)   # Concanate en column
```

```python
df8.head()
```

```python
habbit=df8.smoker.unique()
ordinal=OrdinalEncoder(categories=[habbit])
encoded=ordinal.fit_transform(df8[['smoker']])   
en=pd.DataFrame(encoded,columns=['encoded_smoker'])
df8=df8.drop('smoker',axis=1) 
df8=pd.concat([df8,en],axis=1)  
```

```python
area=df8.region.unique()
ordinal=OrdinalEncoder(categories=[area])
encoded=ordinal.fit_transform(df8[['region']])   
en=pd.DataFrame(encoded,columns=['encoded_region'])
df8=df8.drop('region',axis=1) 
df8=pd.concat([df8,en],axis=1)
```

```python
df8.head()
```
